Agragamos el stock de productos y las ventas mensuales

Sugerir una estrategia de productos a retirar del mercado así como sugerencia de cómo reducir la acumulación de inventario considerando los datos de ingresos y ventas mensuales

```python
historial_sale =[]
tercera_consigna =[]
venta_=0
fecha=[]
stok_venta=[]
fecha_stok_venta=[]
productos_sin_stock=[]
ventas_mayores_al_stock=[]
Productos_en_resago=[]
no_tiene_stok_inicial=[]

lista_ventas=[]
stok_venta=[]
fecha=[]


for producto in lifestore_products:
    stok = producto[4]# Solo toma l primer valor del stok.. despues se le sustrae la venta y queda la diferencia, esto es el indicador que me permite saber cual producto se esta quedando resagado
    for venta in lifestore_sales:
        historial_sale.append(venta + [producto[2]]+ [producto[4]] )
        if venta[1] == producto[0]: # Los id de cada lista.. si coincide el id del producto en las ventas
            venta_+=1
            
            lista_ventas.append(venta_)# una lista de ventas
            stok_venta.append(stok-venta_ )# una lista de diferencias stok-ventas
            fecha.append(venta[3])#una lista de fechas
        else:      
            
            pass
            
       
    if producto[4] !=0:
        
        if venta_> stok: #las ventas superan al stock es decir que puede aumentarse el stok 
            #print('ventas mayores al stock: ',producto,'stok: ',stok,'ventas: ',venta_ )
            ventas_mayores_al_stock.append(( producto[1],['stock_inicial',producto[4]],
                                            ['ventas',lista_ventas],['stock',stok_venta],['fecha',fecha]))
        elif stok==venta_:
            tercera_consigna.append((producto[:2],['stock_inicial',producto[4]],
                                     ['ventas',lista_ventas],['stock',stok_venta],['fecha',fecha]))
            # Estos son todos los productos que no tiene resagos  
            
        elif venta_==0 :
            #print('Productos en resago, ventas menos del 20%', 
            #round((stok-venta_)/producto[4],2),venta_,fecha,producto[1] )
            Productos_en_resago.append((producto[1],['stock_inicial',producto[4]],
                                        ['ventas',lista_ventas],['stock durante el año ',
                                                                 stok_venta],['fecha',fecha]))
        elif (stok-venta_)/producto[4] >= 0.2:
            tercera_consigna.append((producto[:2],['reduccion de stock',(stok-venta_)],['stock_inicial',producto[4]],
                                     ['ventas',lista_ventas],['stock',stok_venta],['fecha',fecha]))
                
        else:
            #estos son los datos que tiene stok y no tiene ventas , son los productos que deben ser retirados de las ventas
            
            tercera_consigna.append((producto[:2],['stock_inicial',producto[4]],
                                     ['ventas',lista_ventas],['stock',stok_venta],['fecha',fecha]))
            # Estos son todos los productos que no tiene resagos 
            
    else:
        #print('no tiene stok :',producto,ventas,fecha)
        no_tiene_stok_inicial.append(( producto[:1],['stock_inicial',producto[4]],
                                      ['ventas',lista_ventas],['stock',stok_venta],['fecha',fecha]))
    venta_=0    
    lista_ventas=[]
    stok_venta=[]
    fecha=[]

```

Elementos donde las ventas han superado el stock, en este esquema es posible aumentarlo a la cantidad de ventas que se realizaron en este año.
